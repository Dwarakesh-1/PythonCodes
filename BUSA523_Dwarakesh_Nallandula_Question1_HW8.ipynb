```python
import pandas as pd
import numpy as np

file = input("Enter the name of the file: ")
filename = file + '.csv'
members = pd.read_csv(filename, parse_dates = ['visitDate','endDate'])
members
```

```python
members_inact = members[members['status'] == 'inactive']
members_inact
```

```python
members_inact.info()
```

```python
members_inact['visitDate'] = pd.to_datetime(members_inact['visitDate'])
members_inact['visitDate']
```

```python
members_inact['endDate'] = pd.to_datetime(members_inact['endDate'])
members_inact['endDate']
```

```python
members_inact['term'] = (members_inact['endDate'] - members_inact['visitDate']).dt.days/30
members_inact
```

```python
male_members = members_inact[members_inact['gender'] == 'M']
female_members = members_inact[members_inact['gender'] == 'F']
```

```python
male_members
```

```python
female_members
```

```python
import scipy.stats as stats

output = stats.ttest_ind(male_members['term'],female_members['term'])
p_value = output.pvalue

print(f"p-value : {round(p_value,4)}")
```

```python
if p_value > 0.05:
    print("The term is not significantly different between men and women")
else:
    print("The term is significantly different between men and women")
```
