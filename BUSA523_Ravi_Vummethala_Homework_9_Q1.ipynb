```python
#a) a stacked bar chart showing the sales of the two divisions over time, both historically and predicted
import pandas as pd

df_IncStmt_A = pd.read_csv('IncStmt-DivA.csv', skiprows=5, usecols=['Description','Act2019','Act2020','Proj2021'])
display(df_IncStmt_A)

df_IncStmt_B = pd.read_csv('IncStmt-DivB.csv', skiprows=5, usecols=['Description','Act2019','Act2020','Proj2021'])
display(df_IncStmt_B)
```

```python
df_A_revenue = df_IncStmt_A[["Act2019","Act2020","Proj2021"]].loc[1]
display(df_A_revenue)

df_B_revenue = df_IncStmt_B[['Act2019','Act2020','Proj2021']].loc[1]
display(df_B_revenue)
```

```python
df_join = df_A_revenue.to_frame().join(df_B_revenue.to_frame(), how='outer', lsuffix='_DivA', rsuffix='_DivB')
display(df_join)
```

```python
df_join.plot(kind='bar', stacked=True, title='Sales over time for DivA & DivB (Stacked chart)')
```

```python
#b) a sales over time clustered bar chart that shows the sales of both divisions   
df_join[["1_DivA", "1_DivB"]].plot(kind="bar", title='Sales over time for DivA & DivB (Clustered bar)')

```

```python
#c) Two distinct pie charts illustrating the different kinds of forecasted expenses for 2021, one for each division



# For Division A
df_A_expenses = df_IncStmt_A[["Description","Proj2021"]].loc[5:9]
df_A_expenses.groupby(['Description']).sum().plot(kind='pie', y='Proj2021', autopct='%1.0f%%', 
                                                  title='Projected 2021 Forecast for Division A')
import matplotlib.pyplot as plt
plt.legend(bbox_to_anchor=(1.25, 1))
plt.show()

# For Division B
df_B_expenses = df_IncStmt_B[["Description","Proj2021"]].loc[5:9]
df_B_expenses.groupby(['Description']).sum().plot(kind='pie', y='Proj2021', autopct='%1.0f%%', 
                                                  title='Projected 2021 Forecast for Division B')
plt.legend(bbox_to_anchor=(1.25, 1))
plt.show()
```
